```python
#Random Data Generated for plotting 
df = pd.DataFrame(np.random.randn(100,3), columns = ['A', 'B', 'C'])
df['A'] = df['A'].cumsum() + 20
df['B'] = df['B'].cumsum() + 20
df['C'] = df['C'].cumsum() + 20
```

```python
df.head()
```

# Line plot

```python
plt.plot(df)
plt.show()
```

* Use the .iplot() method to generate a line plot with the dataset. 

* This plot allows us to click on the elements in the legend to hide and display context which is pretty neat.

* The cursor to the top right of the plot to observe the various features of the plot. We can also use the zoom feature of specific areas of the plot.

```python
df.iplot()
```

```python
df.iplot(x = 'A', y = 'B', mode = 'markers', size = 10)
```

```python
df.iplot(x = 'A', y = 'B', mode = 'lines+markers',colors = 'blue', size = 10)
```

```python
df.iplot(kind = 'scatter',x = 'A', y = 'B', mode = 'markers',colors = 'red', size = 4)
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.iplot(kind = 'bar', x = 'sex', y = 'survived', title = 'Survived', xTitle='Sex', yTitle='Survived')
```

```python
cf.getThemes()
```

```python
cf.set_config_file(theme='henanigans')
```

* barmode determines how bars at the same location coordinate are displayed on the graph. 
* With “stack”, the bars are stacked on top of one another. 

* We can choose barmode = ‘overlay’ to plot the bars over one another for overlapped bar charts, or

* barmode = ‘group’ to place bars beside each other for clustered bar charts 

```python
df.iplot(kind = 'bar', barmode='group', bargap=0.7)
```

```python
df.iplot(kind = 'bar', barmode='stack', bargap=0.7)
```

```python
df.iplot(kind = 'bar', barmode='overlay', bargap=0.7)
```

```python
df.iplot(kind = 'barh', barmode='stack', bargap=0.5)
```

# Histogram

One of 'group', 'overlay' or 'relative' 

In 'relative' mode, bars are stacked above zero for positive values and below zero for negative values. 

In 'overlay' mode, bars are drawn on top of one another. 

In 'group' mode, bars are placed beside each other.

```python
df.iplot(kind='hist', bins = 30, barmode = 'group', bargap=0.5)
```

```python
df.iplot(kind='hist', bins = 30, barmode = 'overlay', bargap=0.5)
```

```python
df.iplot(kind='hist', bins = 30, barmode = 'relative', bargap=0.5)
```

```python
df.iplot(kind = 'box')
```

```python
df.iplot(kind='heatmap',colorscale = 'purd')
```

# Surface Plots: 

Plotly also allows users to plot 3D interactive plots.

```python
#random dataset
df3 = pd.DataFrame({'X': [10,20,30,20,10], 'Y': [10, 20, 30, 20, 10], 'Z': [10, 20, 30, 20, 10]})
df3.head()
```

```python
cf.set_config_file(theme='pearl')
```

```python
df3.iplot(kind='surface', colorscale='rdylbu')
```

The above code displays the numerical features in the dataset along x,y and z-axis in a 3 Dimensional plane.

# datagen

* cufflinks has a special method for generating a random number, called the datagen , for generating random data of different dimensions, such as the following.

```python
cf.datagen.box(50).iplot (kind = 'box', legend = False)
```

See, x-axis has a name corresponding to each box, because the box cufflinks pattern recognition parameter by the kind, it is automatically generated name. If we only generate a random number, it is like this, by default generate random distribution of 100 rows of data, the number of columns selected by themselves.

```python
cf.datagen.histogram(2).iplot(kind = 'histogram')
```

Parameters:         
n_x : int(Number of x categories)
                
n_y : int(Number of y categories)
                                        

```python
cf.datagen.heatmap(20,20).iplot(kind = 'heatmap', colorscale='spectral', title='Cufflinks - Heatmap')
```
