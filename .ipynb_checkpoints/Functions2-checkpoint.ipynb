# Variable Length Argument

```python
# *args:

def Fun1(*arg):
    for i in arg:
        print(i)
```

```python
Fun1(1,2,3,4,5,6,'end')
```

```python
def Fun1(*num):
    for i in num:
        print(i)
```

```python
Fun1(1,2,3,4,5,6,'end')
```

```python
# **kwargs:

def Fun2(**kwargs):
    for i in kwargs.items():
        print(i)
```

```python
Fun2(a=1,b=2,c='Python')
```

```python
def Fun2(**kwargs):
    for i in kwargs.values():
        print(i)
```

```python
Fun2(a=1,b=2,c='Python')
```

```python
# Nested Function:

def fun1():
    print('Python Function')
    def fun2():
        print('inner function')
    fun2()
```

# Practise Questions

```python
# write a function to find the factorial of a number.
def factorial(number):
    fact=1

    for i in range(1,number+1):
        fact = fact*i
    print(fact)
```

```python
factorial(3)
```

```python
# Write a function to check whether number is prime or not.
def prime(num):
    flag = False
    if num ==0 or num ==1:
        print(num,"is not a prime number")
    elif num>1:
        for i in range(2,num):
            if(num % i) == 0:
                flag = True
                break
        if flag:
            print(num,"is not a prime number")
        else:
            print(num,"is a prime number")
```

```python
prime(15)
```

```python
number = int(input('Enter the number'))
count = 0

for i in range(1,number+1):
    if number%i == 0:
        count +=1

if count == 2:
    print(f'{number} is prime number ')
else:
    print(f'{number} is not a prime number')
```

```python
def isprime(number):
    count = 0

    for i in range(1,number+1):
        if number%i == 0:
            count +=1

    if count == 2:
        print(f'{number} is prime number ')
    else:
        print(f'{number} is not a prime number')
```

```python
isprime(3)
```

# List Comprehension:

```python
list1 = [1,2,3,4,5,6]
newlist = []

for i in list1:
    newlist.append(i*2)

print(newlist)
```

```python
list1 = [1,2,3,4,5,6]

newlist = [i for i in list1]
print(newlist)
```

```python
# List Comprehension:
list1 = [1,2,3,4,5,6]

newlist = [i*2 for i in list1]
print(newlist)
```

```python
# Generate a list from 1 to 50 using list comprehension:

list2  = [x for x in range(1,50)]
print(list2,end='')
```

```python
print(list2)
```

```python
# if condition in List Comprehension

list3 = [23,44,54,66,77,88,1,2,3,4]

nlist = [i for i in list3 if i>50]
```

```python
print(nlist)
```
